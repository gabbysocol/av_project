```python
import pandas as pd
import numpy as np
import re

import warnings
warnings.filterwarnings('ignore')
```

```python
def save_df(df, file_name):
    df.to_csv(file_name, sep=';')
```

```python
def read_file(file_name):
    df = pd.read_csv(file_name, sep=';')
    pd.set_option('display.max_columns', 125)
    return df
```

```python
def preproc_name(df):
    i = 0
    while i < len(df):
        match = re.search(r'[A-Za-z\-]+[\ ]+[A-Za-z\-А-Яа-я0-9]*', str(df['Название'][i]))
        if match:
            df['Название'][i] = match[0]
        i += 1
    return df
```

```python
def preproc(avto):
#     del avto['Unnamed: 0']

    avto = avto.loc[avto['Название'] != 'None']
    
    avto = preproc_name(avto)
    
    avto['Цена'] = avto['Цена'].str.replace('р', '').astype('float')
    
    avto['Год'] = avto['Год'].fillna('0').str.replace(r'[А-Яа-яA-Za-z\ ]*', '').astype('int')
    
    avto['Состояние'] = avto['Состояние'].fillna('0').str.replace(r'Состояние\ *', '').astype('str')

    avto['Тип топлива'] = avto['Тип топлива'].fillna('0').str.replace(r'Тип топлива\ ', '').astype('str')

    avto['Объем'] = avto['Объем'].fillna('0').str.replace('Объем ', '').replace(' см3', '').astype('int')

    avto['Пробег'] = avto['Пробег'].str.replace('Пробег ', '')
    avto['Пробег'] = avto['Пробег'].str.replace(' км', '')
    miles_mask = avto['Пробег'].str.endswith(' миль')
    avto.loc[miles_mask, 'Пробег'] = avto[miles_mask]['Пробег'].str.replace(' миль', '').astype('int') * 1.60934
    avto['Пробег'] = avto['Пробег'].astype('int')
    
    avto['Цвет'] = avto['Цвет'].fillna('0').str.replace(r'Цвет\ *', '').astype('str')
    
    avto['Тип кузова'] = avto['Тип кузова'].fillna('0').str.replace(r'Тип кузова\ *', '').astype('str')
    
    avto['Трансмиссия'] = avto['Трансмиссия'].fillna('0').str.replace(r'Трансмиссия\ *', '').astype('str')
    
    avto['Привод'] = avto['Привод'].fillna('0').str.replace(r'Привод\ *', '').astype('str')

    return avto
```
